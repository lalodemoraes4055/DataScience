Recuerda que:


*   Una **clave primaria** es un atributo (o conjunto) que identifica unívocamente a cada registro en la tabla.
*   Una **clave foránea** (externa o ajena) es un atributos (o conjunto) en una tabla que es una clave primaria en otra (o posiblemente la misma) tabla.
*   Las **relaciones** son las líneas que conectan una tabla con otra y el extremo determina la cardinalidad. Las relaciones con línea continua (identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser foránea y primaria (al mismo tiempo) de otra. Las relaciones con línea discontinua (no identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser sólo foránea en otra.

8.	La cantidad de productos por línea de producto (no las existencias en inventario)

9.	La cantidad de empleados por país (tomando en cuenta la ubicación de la oficina).

```python
# Consulta de la cantidad de empleados por país
query = text("SELECT o.country, COUNT(e.employeeNumber) AS employeeCount FROM employees e JOIN offices o ON e.officeCode = o.officeCode GROUP BY o.country")
df8 = pd.read_sql(query, conn)
df8
```

Se utiliza la información de employees y offices, en este caso tambien se generan abreviaviones o alias para no repetir el nombre completo de la base de datos, por ello se utiliza un join y con un group by se separa por país

10.	El promedio de los pagos de cada uno de los clientes de España (sin incluir aquellos que no poseen ningún pago).

```python
# Consulta del promedio de pagos de los clientes en españa
query = text("SELECT c.customerName, AVG(p.amount) AS averagePayment FROM customers c JOIN payments p ON c.customerNumber = p.customerNumber WHERE c.country = 'Spain' GROUP BY c.customerName")
df9 = pd.read_sql(query, conn)
df9
```

Se utilizan 2 db customers y payments igual utilizando abreviaciones, incorporando la función de AVG, se utiliza un join, se filtra por el país de España y finalmente se agrupan por el nombre del cliente.

11.	Carga las tablas empleadas en dataframes con el mismo nombre y resuelve las consultas anteriores con las funciones de Pandas (NO con SQL). Cuida no sobreescribir los dataframes originales al resolver las consultas. Debes obtener los mismos resultados que con SQL.

```python
productlines = pd.read_sql(text("Select * FROM productlines"), conn)
customers = pd.read_sql(text("Select * FROM customers"), conn)
products = pd.read_sql(text("Select * FROM products"), conn)
employees = pd.read_sql(text("Select * FROM employees"), conn)
offices = pd.read_sql(text("Select * FROM offices"), conn)
payments = pd.read_sql(text("Select * FROM payments"), conn)
```

1. La información de las líneas de productos.

```python
# Consulta para obtener las líneas de productos
productlines
```

2. La información de los empleados ordenados por nombre (firstName).

```python
# Consulta para la información de los empleados ordenados por el primer nombre
employees.sort_values(by='firstName')
```

3. Los países donde hay oficinas (sin duplicar).

```python
# Consulta de países donde hay oficinas
offices['country'].drop_duplicates()
```

4. El nombre y teléfono de los clientes de la ciudad de Nueva York (NYC).

```python
# Consulta de nombres y telofonos de clientes de NYC
customers[customers['city'] == 'NYC'][['customerName', 'phone']]
```

5. El código y nombre de los productos del vendedor Gearbox Collectibles que tengan menos de 1000 unidades en stock.

```python
# Consulta de productos de Gearbox que tienen menos de 1000 unidades
products[(products['productVendor'] == 'Gearbox Collectibles') & (products['quantityInStock'] < 1000)][['productCode', 'productName','quantityInStock']]
```

6. Los tres productos más caros, desde el punto de visto de los comercializadores (buyPrice).

```python
# Consulta de los productos más caros
products.sort_values(by='buyPrice', ascending=False)[['productCode', 'productName', 'buyPrice']].head(3)
```

7. La cantidad de productos por línea de producto (no las existencias en inventario)

```python
# Consulta de la cantidad de productos por línea de producto
productlines.merge(products.groupby('productLine').size().reset_index(name='productCount'), on='productLine')[['productLine', 'productCount']]
```

8.La cantidad de empleados por país (tomando en cuenta la ubicación de la oficina).

```python
# Consulta de la cantidad de empleados por país
employees.merge(offices[['officeCode', 'country']], on='officeCode').groupby('country').size().reset_index(name='employeeCount')
```

9. El promedio de los pagos de cada uno de los clientes de España (sin incluir aquellos que no poseen ningún pago).

```python
# Consulta del promedio de pagos de los clientes en españa
payments.merge(customers[customers['country'] == 'Spain'][['customerNumber', 'customerName']], on='customerNumber').groupby('customerName')['amount'].mean()
```
